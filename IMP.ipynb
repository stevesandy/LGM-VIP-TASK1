## Task-1: Image to Pencil Sketch with Python

```python
#importing libraries

import cv2
from matplotlib import pyplot as plt
```

```python
#reading the image

image = cv2.imread("bird.jpg")
color = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (10,10))
plt.axis("off")
plt.imshow(color)
plt.title('Original Image')
plt.show
```

```python
#converting the original image to greyscale

grey_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize = (10,10))
plt.axis('off')
plt.title('GreyScale Image')
plt.imshow(grey_img,cmap='gray')
plt.show()
```

```python
#inverting the new grey scale image
inverted_img = 255 - grey_img
plt.figure(figsize = (10,10))
plt.axis('off')
plt.title('Inverted Image')
plt.imshow(inverted_img,cmap='gray')
plt.show()
```

```python
#Blurring the image using the Gaussian Function

blurred = cv2.GaussianBlur(inverted_img, (21,21),0)
plt.figure(figsize = (10,10))
plt.axis('off')
plt.title('Blurred Image')
plt.imshow(blurred,cmap='gray')
plt.show()
```

```python
#inverting the Blurred image

invert_blurred = 255 - blurred
pencil = cv2.divide(grey_img,invert_blurred,scale=256.0)
plt.figure(figsize = (10,10))
plt.axis('off')
plt.title('Pencil Image')
plt.imshow(pencil,cmap='gray')
plt.show()
```

```python
#Comparing the original image and the new pencil image

plt.figure(figsize = (10,10))
plt.subplot(1,2,1)
plt.imshow(color)
plt.axis('off')
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(pencil,cmap='gray')
plt.axis('off')
plt.title('Pencil Image')
```
